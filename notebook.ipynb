## Регрессионный анализ - предсказание одной переменной на основе другой(линейная регрессия) - предсказывает, корреляция - аналзирует взаимосвязи

В регрессии - *зависимую* переменную пытаемся научиться предсказат, а *независимые* для этого используем

Одним из простеших методов описать, как зависимая переменная влияет на независимую - провести линию, которая определенным образом усредняа бы все знчения.

Как известно обычную линию можно описать уравнением

$y = b_o + b_1x$ , где $b_o$ - линия пересечиня с осью y - intecept, а $b_1$ и последующие коэффициенты - slope(показывает направление и наклон линии). Чтобы подобрать эти праметра существует метод наименьших квадратов

### Метод наименьших квадратов
Метод нахождения оптимальных параметров линейной регрессии, таких что сумма квадратов *ошибок* была минимальной.

Ошибка описывается расстоянием от реального значения величины, до его проекции на линию регресси, соотвестветнно, чем это расстояние меньше - тем лучше.

МНК помогает подобрать коэффициенты следующим образом

$b_1 = \frac{sd_y}{sd_x}r_xy$

$b_0 = \overline{Y}- b_1\overline{X}$

Чтобы ответить на вопрос насколько статистически значима связь переменных нужна $H_0 - b_1 = 0$ тоесть коэффициент корреляции равен 0 и линия регрессии будет параллельна оси х со сзначением среднего переменной Y.

Если мы много раз будем проводить эксперимент, предполагая, что $b_1 = 0$ то выборочные значения коэффициентом $b_1$ распределились бы нормально вокруг нуля, и мы можем расчитать т-критерий

$t = \frac{b_1 - 0}{se}$ , проверяем гипотезу о том, что в ГС $b_1$ отличен от нуля.

## Нормальное распределине

Распределние как правило унимодально и симетрично, но более важно что отклоенние значений от среднего подчиняются определенному веротяностному закону

*Стандартизация* - преобразование данных в стандартную шкалу, где среднее

$M_z$ = 0, $D_z = 1$

Для того, чтоб перевести значения в z-value необходимо из каждого значения вычесть среднее, а затем поделить на sd по выборке, чтобы отнести изменение величины от среднего относительно стандарта

$Z = \frac{x_i - \overline{x}}{sd}$

### Центральная предельная теорема

$se = \frac{σ}{\sqrt{n}}$
Однако если выборка является репрезентативной, то мы имеем право применить следующую формулу
$se = \frac{sd_x}{\sqrt{n}}$

Таким образом, представим ситуацию, что мы сделали лишь одну выборку и смогли расчитать стандартную ошибку среднего, тогда все выбороынчые средние буду отклоняться от некоего среднего ГС на вычисленную нами se

Из ГС делаем несколько выборок и анализируем их отклоенния и среднее и видим, что они разнятся, однако если распределить сами средние значения, то получим результат схожий со средним в ГС(стандартное отклонение этих распределений - стандартная ошибка среднего - se) и чем больше элементов в выборке, тем точнее стандартная ошибка среднего.

## T - распределние
Если число наблюдений невелико и σ неизвестно используется распределние Стьюдента.
Основным отличием является его неточность при маленьких степенях свободы. Под хвостами распределения будет лежать больше наблюдений, а при отклонении на 1 σ меньше, так как график будет ниже. Однако при увеличении степени свободы, т-распределние будет стремится к нормальному.

Грубо говоря, т - распределние используется при n < 30, однако на самом деле мы используем его из-за того, чт не знаем σ в ГС. Находить отклонение от среднего в ГС будет правильнее в терминах т - распределиня, поэтому впредь для такой задачи используется т-распределние.

df - число степеней свободы = (n - 1) зависит от числа наблюдений в выборке.
Критерий т - Стьюдента помогает сравнить два выборочных средних

## Сравнение два средних значений выборки

Формулируем две гипотезы - $H_0 => \mu_1 = μ_2$ , тоесть если средние значения выборок одинаковы, то и средние в ГС будут одниаковы, это говорит о том, что в выборках нет различий, значит на них не повляил никакой фактор(например лекарство не сработало на вторую выборку и они оказались такими же как первая)

$H_1 => \mu_1 \neq μ_2$

Проверяем первую гипотезу - если мы предположили, что средние значения равны, то их разность была бы равна $\overline{X_1} - \overline{X_2} = \overline{\mu_1} - \overline{\mu_2} = 0$

В данном случае ошибка среднего se расчитывается по формуле $se = \sqrt{\frac{sd^2_1}{n_1} + \frac{sd^2_2}{n_2}}$

При большом числе наблюдений разница срдених распределится нормально, но еще более точно будет сказать, что она распределится по т - распределиню с число степней свобод $df = n_1 + n_2 - 2$

Теперь мы можем расчитать насколько конкретная разность средних по выборке отклонилась от среднего ГС.

$t = \frac{(\overline{X_1} - \overline{X_2}) - (\overline{\mu_1} - \overline{\mu_2})}{se} = \frac{(\overline{X_1} - \overline{X_2}) - (\overline{\mu_1} - \overline{\mu_2})}{\sqrt{\frac{sd^2_1}{n_1} + \frac{sd^2_2}{n_2}}}$

 для первой гипотезы разность средни в ГС равны, поэтому формула примет следующий вид:

$t =  \frac{(\overline{X_1} - \overline{X_2})}{\sqrt{\frac{sd^2_1}{n_1} + \frac{sd^2_2}{n_2}}}$

## Регрессионный и корреляционные анализы

Корреляция можеь измерить только линейную связь.

Если коэффициент корреляции равен 0, то либо связи нет между переменными, либо между ними нелинейная связь.

### Коэффициенты корреляции

Ковариация - показьель корреляции, корреляционный момент

$cov(x,y) = ∑\frac{(x_i-\bar{x})⋅(y_i-\bar{y})}{n}$

определяется как совместный центральный омент второго порядка или матожидание произведения отклонени СВ от своих матожиданий. Если положительный, то ряды между собой связаны положительно, если отрицательный, то метрики в разные стороны двигаются.

Коэффициент Пирсона

*Нормированный*, параметрический показатель. В расчете используется не только разница, но еще нормируется на скв.

$r_{XY} = \frac{cov_{XY}}{σ_{X}σ{Y}}=∑\frac{\sum{(X-\bar{X})(Y-\bar{Y})}}{\sqrt{\sum{(X-\bar{X})^2(Y-\bar{Y})^2}}}$

Лежит между [-1;1] все логично.

Коэф пирсона действует только если связь линейная, во - вторых он очень чувствителен к *выбросам*. Если не убрали выбросы, то плохой результат будет.

МОжно применять только если ряды не имеют ярко выраженной ассиметрии, если данные количественные, а не качественные, нет выбросов, связь прямолинейная.

Коэффициент Кендаллла и Спирмена - не параметрические, могут быть использоваться, если предыдущие не подошли в основе лежит ранжирование.

Спирмен - формулу лучше загуглить. Значит ранжируем первую переменную, вторая в соответствтии с ней ранжируется потому по формуле считаем щит и в конце получаем коэфф.

Коэффициент Кендалла $r=\frac{2(\sum{P}-\sum{Q})}{n(n-1)}$, где P - сумма количества показателей ряда у которых ранг Y больше, а Q тоже самое но ранг меньше.

Начинаем также как в Спирмене, ранжируем.

## Регрессионный анализ

$\widehat{y} = a+b_1⋅x_1+b_2⋅x_2 +...+ b_n⋅x_n+ε$ - множественная регрессия

* сперва определяем тип взаимосвязи, не сразу ищем коэффициенты
* затем оцениваем параетры уравнения регрессии
* затем проверяем степень значимости уравнения регрессии и его степень точности
* построить интервальную оценку прогноза, не только точечную, потому что всегда есть ошибка и шанс попасть в точечную мал
* исследовать характеристики точности модели
* построить точечные и интервальные прогнозы результрующей переменной

Самый распространенный вид регрессии - линейная (*множественная* и *парная*)

$\widehat{y} = a+b⋅x+ε$

Для того, чтобы определить насколько точно есть метрики, но графически лучше всего линией (не всегда бывает линейная, бывает экспонинцеальная, степенная и тд)

Бывает когда процесс происходит рывками. Тогда независимая переменная х может быть разбита на классы или сегменты и может быть осуществлена линейная регрессия посегментно.

В основе лежит метод наименьших квадратов. Ищем такие параметры с помощью которых разница между фактически значений и расчетными(проекции их на линию регрессии) будет стремиться к миниуму.

План действий при проведении РА

Во первых определяемся с факторами, которые будем включать в модель. Чем больше факторов, тем больший врменной ряд необходим, чтобы модель прогноза была качественной.

Определяем параметры уравнения регрессии a и b

Подставляем значения в уравнения.

Чобы проверить значимость уравнения регрессии есть метрики.

* коэффициент детерминации - показывает какой процент изменения нашей ключевой метрики описывается *именно данным* уравнением регрессии.

$R^2 = 1 - \frac{\sum{(y_i-\widehat{y}_i)^2}}{\sum{(y_i-\bar{y}_i)^2}}$

$R^2 = 1 - \frac{D_{res}}{D_{total}}$ , где $D_{res}$ - сумма квадратов остатков (сумма разностей реального значения и его координаты на линии регрессии в кавадртае), а $S_{total}$ - общая изменчивость - сумма расстояний в координатах от наблюдения до среднего значения. Это единица минус доля необъясненной дисперсии.

Не может быть отрицательным, чем ближе к 1 тем более точное и больше факторов включили. Должен быть > 0.5

Критерий Фишера

Коэффициент детерминации моз=жет применяться для проверки гипотезы о значимости уравенния регрессии.

Для этого рассчитвают статистику $F_{набл}$ имеющую распределение Фишера:

$F_{набл} =\frac{\sum{(\widehat{y}_i-\bar{y}_i)^2/1}}{\sum{(y_i-\widehat{y}_i)^2/(n-2)}}$

$F_{набл} = \frac{R^2}{1-R^2}⋅(n-2)$

Получаем значение и сравниваием его с критическим значением для опредленного значений степеней свободы. И нулевая гипотеза говорит о том что расчетное значение меньше критического. ПОэтому если реально оно больше, то мы отвергаем гипотезу о том что уравнение не значимо.

Регрссионый анализ легкий, но не совсем точный, потому что экстраполируем. Плюс человек сам принимая решение может ошибиться и не включает все факторы в рассмотрение от этого не очень точно.

## Проверка статистических гипотез

Как правило $H_0$ изначально говорит, что эффекта нет.
И если мы ее отклоняем, значит есть основания.

Сперва всегд азадаем критерий значимости.

Односторонний критерий чаще рассматривает в контексте меньше - больше, а двусторонний сравнивается с определенным значением.

p-value - вероятноть получить наблюдаемое или еще большее отклонение оценки от гипотезы, если она верна.

ПО сути это вероятность отвергнуть нулевую гипотезу при условии что она верна.

Если p-value < α(уровня значимости альфа) то нулевую гипотезу можно отвергнуть и принять альтернативную гипотезу.

Пи-вэлью - вероятность получить такие ли еще более выраженные различия выбороки при условии что в ГС различий нет, тоесть что H0 врна.

### критерий Пирсона или $χ^2$

критерий хи-квадрта применяется для категориальных данных (класс автомобиля, пол участника эксперимента)

Для расчета наблюдаемые частоты обозначаются через О(observed) и E(expected)

$χ^2_n = ∑\frac{(O_i-E_i)^2}{E_i}$

Например частоты выпадения очков на игральной кости, ожидаем столько-то получаем столько-то  и определяем не обманывают ли нас.

Получив значений сравниваем с табличным.

Если расчетное значений критерия меньше табличного, то гипотеза не отклоняется.

ПО графику видно, что наше значение критерия не попадает в критическую область, а значит и вероятность получить такую оценку выше критического значеиня.


### A/B тесты

Switchback - тестирование снижает риск подобных сетевых эффектов, рандомизируя эксперимент в батчи по регионам и времени суток, а не попользователям или объектам рекомендации.

A/B тесты нужны когда правильный ответ на основве имеющихся данных сформулировать невозможно, когда не знаете как быть, когда надо понять, а не придумать как ведут себя польщователи или рынок.

## определение выбросов
 с помощью стандартного отклонения

```python
import numpy as np
import matplotlib.pyplot as plt


# multiply and add by random numbers to get some real values
data = np.random.randn(50000)  * 20 + 20

# Function to Detection Outlier on one-dimentional datasets.
def find_anomalies(data):
    #define a list to accumlate anomalies
    anomalies = []

    # Set upper and lower limit to 3 standard deviation
    random_data_std = np.std(data)
    random_data_mean = np.mean(data)
    anomaly_cut_off = random_data_std * 3

    lower_limit  = random_data_mean - anomaly_cut_off
    upper_limit = random_data_mean + anomaly_cut_off
    print(lower_limit)
    # Generate outliers
    for outlier in data:
        if outlier > upper_limit or outlier < lower_limit:
            anomalies.append(outlier)
    return anomalies
find_anomalies(data)
```

  Квантификатор после символа или группы определяет, сколько раз предшествующее выражение может встречаться.

    {n,m}  повторений может быть от n до m включительно.
    {n,}   n и более повторений.
    {,m}   не более m повторений.
    {n}    ровно n повторений

Методы и функции, предназначенные для работы с регулярными выражениями, не входят в базовый функционал языка Python. Тем не менее эта проблема легко разрешима. При разработке программ на Python мы можем, помимо использования стандартной библиотеки, подключать дополнительные библиотеки и модули, тем самым расширяя свои возможности по созданию программ. Для Python написано огромное количество модулей и библиотек.

```python

import re
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
dict_ = {}
for elem in emails:
  res = re.findall(r'\b\w+@(\w+.\w+)\b',elem)
  print(res[0])
  if dict_.get(res[0]):
    dict_[res[0]] += 1
  else:
    dict_[res[0]] = 1
print(dict_)

```

```python
some_text = '''
Эталонной реализацией Python является интерпретатор CPython,
поддерживающий большинство активно используемых платформ.
Он распространяется под свободной лицензией Python Software Foundation License,
позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.
'''
cons = 0
vowels = 0
res = re.split(r'[\s,]+', some_text.strip())
low = list(map(lambda x : x.lower(),res))
print(low)
for word in low:
  if word.startswith(('а','е','и','о','у','ё','ю','я','a', 'e', 'i', 'o', 'u')) : vowels+=1
  else: cons += 1

print(cons, vowels)
```

```python
import re
from re import IGNORECASE
some_text = '''
Эталонной реализацией Python является интерпретатор CPython,
поддерживающий большинство активно используемых платформ.
Он распространяется под свободной лицензией Python Software Foundation License,
позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.
'''
res = re.findall(r'\b[АЕЁИОУЫЭЮЯAEIOU]', some_text.strip(), flags=re.IGNORECASE)
print(len(res))
```
